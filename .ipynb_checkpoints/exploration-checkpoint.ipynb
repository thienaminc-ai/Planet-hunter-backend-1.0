```python
import pandas as pd
import numpy as np

# Load file
df = pd.read_csv('data/kepler_data.csv', comment='#')

# Kiểm tra NaN và phân phối
print("Số lượng NaN trong koi_kepmag:", df['koi_kepmag'].isnull().sum())
print("Thống kê koi_kepmag:", df['koi_kepmag'].describe())

# Mã hóa nhãn (nếu cần)
df['koi_disposition'] = df['koi_disposition'].map({'CONFIRMED': 0, 'CANDIDATE': 1, 'FALSE POSITIVE': 2})
print("Tương quan với koi_disposition:", df['koi_kepmag'].corr(df['koi_disposition']))
```

```python
import pandas as pd
import numpy as np

# Load file CSV
file_path = 'data/kepler_data.csv'
df = pd.read_csv(file_path, comment='#')

# Kiểm tra trường có toàn 0
zero_columns = (df == 0).all()
print("Các trường có toàn bộ giá trị = 0:")
print(df.columns[zero_columns].tolist())

# Kiểm tra trường có NaN
nan_columns = df.isnull().any()
print("\nCác trường có giá trị NaN:")
print(df.columns[nan_columns].tolist())

# Thống kê số lượng NaN mỗi cột
print("\nSố lượng NaN mỗi cột:")
print(df.isnull().sum())
```

```python
import seaborn as sns

# Vẽ biểu đồ phân phối chu kỳ quỹ đạo
sns.histplot(data=df, x='koi_period')
```
